```python
import numpy as np
```

```python
# vector length
v1 = np.array([1,2,3,4,5,6])

vl1 = np.sqrt(sum(np.multiply(v1,v1)))
vl2 = np.linalg.norm(v1)

print(vl1,vl2)
```

```python
# dot product based on theta:
# alpha as representing dot product
# cos(theta) > 0 => alpha > 0 (theta less than 90*, acute)
# cos(theta) < 0 => alpha < 0 (theta greater than 90*, obtuse)
# cos(theta) = 0 => alpha = 0 (theta = 90*) -- ~~~ORTHOGONAL~~~
# cos(theta) = 1 => alpha = |a||b| (theta = 0*) -- DEPENDENT
```

# 

```python
# the proof for algebraic and geometric interpretation being the same is based on the law of cosines C:

v1 = np.array([2,4,-3])
v2 = np.array([0,-3,-3])

# compute angle in radians b/w 2 vectors
theta = np.arccos(np.dot(v1,v2)/(np.linalg.norm(v1)*np.linalg.norm(v2)))
print(theta)
```

```python
import matplotlib.pyplot as plt

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot([0,v1[0]],[0,v1[1]],[0,v1[2]],'b')
ax.plot([0,v2[0]],[0,v2[1]],[0,v2[2]],'r')

plt.axis((-6,6,-6,6))
plt.show()
```

```python
# python demonstration of algebraic and geometric dot prod~

dotprod_alg = np.dot(v1,v2)
dotprod_geom = np.linalg.norm(v1)*np.linalg.norm(v2)*np.cos(theta)

print(dotprod_alg)
print(dotprod_geom)
```
